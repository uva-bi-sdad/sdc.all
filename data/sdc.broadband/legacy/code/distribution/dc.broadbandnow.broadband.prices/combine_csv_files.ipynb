```python
import pandas as pd
import os
```

Because files are more reliably downloaded piecemeal, we need to recombine it

```python
d = '../../data/reverse_address'

dfs = []
for file in sorted(os.listdir(d)):
    if file.split('.')[-1] == 'csv':
        dfs.append(pd.read_csv(os.path.join(d, file)))

df = pd.concat(dfs)
```

```python
df.to_csv(d+'county_addresses.csv', index=False)
```
