# Fix combine malformed files

```python
dfs = []
for file in tqdm(to_parse):
    filepath = os.path.join('../data/temp', file)

    with open(filepath, 'r', encoding='utf-8') as file:
        data = file.readlines()
        
    data[0] = "index,input,match,non_exact,street,coordinate,tiger,lr\n"

    with open(filepath, 'w', encoding='utf-8') as file:
        file.writelines(data)
    df = pd.read_csv(filepath, dtype={'tiger': object, 'coordinate': object})
    df = df[['match', 'non_exact', 'street', 'coordinate']]       
    dfs.append(df)
tdf = pd.concat(dfs)
tdf= tdf[tdf['coordinate'].notna()]
tdf    
```

```python
def s2p(s):
    try: 
        return (s.split(',')[0], s.split(',')[1])
    except Exception as e:
        return e
```

```python
tdf['coordinate'] = tdf['coordinate'].apply(lambda x: s2p(x))
```

```python
tdf = tdf[tdf['coordinate'].notna()]
tdf
```

```python
tdf['lon'] = tdf['coordinate'].apply(lambda x: x[0])
tdf['lat'] = tdf['coordinate'].apply(lambda x: x[1])
tdf
```

```python
tdf = tdf[['street', 'lon', 'lat']]
tdf
```

```python
tdf.to_csv('../data/temp/01073_points_address.csv', index=False)
```
